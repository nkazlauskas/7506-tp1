```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
events = pd.read_csv('events.csv.gzip', compression = 'gzip', low_memory=False)
events
```

```python
clicks = pd.read_csv('clicks.csv')
```

#### Buscamos los devices que se encuentran tanto en el dataset con información sobre los clicks como en el de información sobre los events

```python
events_clicks = pd.merge(clicks, events, on='ref_hash', how='inner')
```

```python
len(events_clicks)
```

```python
events_clicks
```

#### Queremos ver cuales son los eventos que mas ocurrencias tienen en estos usuarios, que tb tienen un click

```python
top_10_events = events_clicks['event_id'].value_counts().sort_values(ascending = False).head(10)
top_10_events
```

```python
top_10_events.sum()/len(events_clicks['event_id'])
```

#### Los 10 eventos con más ocurrencias representan el 83,14% del total de la muestra

```python
plt.figure(figsize=(12, 6))
sns.barplot(x = top_10_events.index, y = top_10_events.values, palette = sns.color_palette("Spectral", 10))

plt.xticks(rotation = 0, fontsize = 'medium')
plt.tick_params(top = False, bottom = False, left = False, right = False, labelleft = True, labelbottom = True)
plt.ylabel('Cantidad de eventos', labelpad = 20, fontsize = 'large')
plt.title('Cantidad de eventos para cada tipo de evento en usuarios con clicks', fontsize = 'xx-large')
plt.xlabel('Tipo de evento', fontsize = 'large')
plt.show();
```

```python
#Timestamp en que fue creado el evento
events_clicks['event_timestamp'] = pd.to_datetime(events_clicks.date)

#Timestamp en que fue creado el evento de click
events_clicks['clicks_timestamp'] = pd.to_datetime(events_clicks.created)

#Calculamos la diferencia entre un evento y un evento de click, en segundos
events_clicks['clicks_delta'] = (events_clicks['event_timestamp'] - events_clicks['clicks_timestamp']).dt.total_seconds()
```

#### Queremos ver cómo es la distribución de los eventos alrededor del evento del click


```python
plt.figure(figsize=(12, 6))
plt.hist(events_clicks['clicks_delta'], bins = 50, color = '#008751', alpha = 0.8)

plt.tick_params(top = False, bottom = True, left = False, right = False, labelleft = True, labelbottom = True)
plt.axvline(0, color='#40434E', linewidth=1.5, linestyle='dashed')
plt.ylabel('Cantidad de eventos', labelpad = 20, fontsize = 'x-large')
plt.xlabel('Diferencia temporal entre clicks y evento (segundos)', labelpad = 20, fontsize = 'x-large')
plt.title('Cantidad de eventos en función de la diferencia temporal entre evento y clicks', fontsize = 'xx-large', pad = 20)
plt.show()
```

#### Vemos que hay ocurrencia de eventos pre y post-click. La mayor cantidad de ocurrencias sucede muy cercana al click.

#### Nos interesa saber cuáles son esos eventos

```python
events_clicks['pre_click'] = events_clicks['clicks_delta'] < 0
```

```python
events_pre_click = events_clicks[events_clicks['pre_click'] == True]
events_pre_click = events_pre_click['event_id'].value_counts().head(10)
events_pre_click
```

```python
events_post_click = events_clicks[events_clicks['pre_click'] == False]
events_post_click = events_post_click['event_id'].value_counts().head(10)
events_post_click
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (22, 6))
plt.subplots_adjust(hspace = 0.4)

ax1.bar(range(10), events_pre_click.values, color = sns.color_palette('OrRd_r', 10))
ax1.set_xticks(range(10))
ax1.set_xticklabels(events_pre_click.index)
ax1.set_xlabel('Eventos pre-click', fontsize = 'x-large')
ax1.set_ylabel('Cantidad de eventos', fontsize = 'x-large')

ax2.bar(range(10), events_post_click.values, color = sns.color_palette('YlGn_r', 10))
ax2.set_xticks(range(10))
ax2.set_xticklabels(events_post_click.index)
ax2.set_xlabel('Eventos post-click', fontsize = 'x-large')

fig.suptitle('Ubicación temporal de eventos mayoritarios con respecto al click', fontsize = 'xx-large');
plt.show();
```

#### Vemos que existen eventos que se encuentran en ambos períodos, mientras que otros pertenecen únicamente a uno


```python
pre = set(events_pre_click.index)
post = set(events_post_click.index)
```

```python
in_both = pre.union(post)
only_pre = pre.difference(post)
only_post = post.difference(pre)
```

#### Se pude observar que no parece haber una gran diferenciación significativa entre los eventos pre y post click. Nos disponemos a analizar los 2 eventos con mayores cantidades de eventos para ambas categorías

```python
event_22 = events_clicks[events_clicks['event_id'] == 22]['clicks_delta']
event_31 = events_clicks[events_clicks['event_id'] == 31]['clicks_delta']

plt.figure(figsize=(16, 8))
plt_params ={'hist':False, 'kde_kws': {'bw': 0.15}, 'norm_hist': False}
sns.distplot(event_22, label = "Evento 22", **plt_params)
sns.distplot(event_31, label = "Evento 31", **plt_params)
plt.axvline(0, linestyle = '--', color = 'grey')
plt.tick_params(top = False, bottom = False, left = False, right = False, labelleft = True, labelbottom = True)
plt.ylabel('Densidad', labelpad = 20, fontsize = 'large')
plt.title('Estimación de densidad de eventos en función de la diferencia temporal entre evento y click ', fontsize = 'xx-large')
plt.xlabel('Diferencia temporal entre click y evento (segundos)', fontsize = 'large')
plt.legend(frameon = False, loc = 1, handletextpad = 0.9, fontsize = 'large')
plt.show();
```
